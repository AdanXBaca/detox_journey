```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def R_0(epsilon, beta, b, sigma, gamma, eta, mu):
  return R_01(epsilon, beta, b, sigma) + R_02(beta, b, gamma, eta, mu)

def R_01(epsilon, beta, b, sigma):
  return (epsilon * beta * (1-b)) / sigma

def R_02(beta, b, gamma, eta, mu):
  return (beta * (1-b)) / (gamma + eta + mu)
```

```python
# beta = 0.5
epsilon = 0.9
b = 0
sigma = 0.9
gamma = 1/14
eta = 1/21
mu = 0.05

beta_R0_vals = []
betas = np.linspace(0, 1, 1000)
for beta in betas:
  beta_R0_vals.append(R_0(epsilon, beta, b, sigma, gamma, eta, mu))
  # print(beta)
  # print(R_0(epsilon, beta, sigma, gamma, eta, mu))
plt.plot(betas, beta_R0_vals)
plt.xlabel(r"$\beta$ value")
plt.ylabel(r"$\mathcal{R}_0$")

```
